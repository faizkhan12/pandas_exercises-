```python
import matplotlib.pyplot as plt
```

```python
% matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'y')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot')
```

```python
#Multiplot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x)
```

```python
# Object oriented Approach
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'y')
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('TITLE')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("AAAA")
axes2.set_title('BBB')
```
